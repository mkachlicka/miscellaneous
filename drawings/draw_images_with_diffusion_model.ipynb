```python
# Load packages
from diffusers import DiffusionPipeline
import torch
```

```python
# Draw something using pretrained diffuser model
# It takes a few minutes to run, especially without cuda your laptop might get loud and hot!
pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float32)
#pipeline.to("cuda") # Check if cuda available (see below)
pipeline("An image of a squirrel in Picasso style").images[0]
```
